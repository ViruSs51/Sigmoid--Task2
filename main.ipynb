**Deschiderea la dataset**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('wine-quality-white-and-red.csv')

df
```

**Transferarea din string in integer**

```python
count = df['type'].value_counts()

if 'white' in count and 'red' in count:
    df['type'] = df['type'].map({'white': 0, 'red': 1})

count
```

*Parametrile principale a datasetului*

```python
for axis in range(2):
    print(f'Media{axis+1}', df.mean(axis=axis), sep=' ---> ', end='\n\n')
    print(f'Mediana{axis+1}', df.median(axis=axis), sep=' ---> ', end='\n\n')
    print(f'Modul{axis+1}', df.mode(axis=axis), sep=' ---> ', end='\n\n\n')
    print(f'Std{axis+1}', df.std(axis=axis), sep=' ---> ', end='\n\n\n')

print('\n', df.describe(), sep='')
```

```python
def max_correlation(dataframe: pd.DataFrame, column_name: str):
    column_correlation = np.sort(dataframe.corr()[column_name].abs())
    
    return [column_correlation[-c] for c in range(1, 4)]

max_correlation(df,'pH')
```
